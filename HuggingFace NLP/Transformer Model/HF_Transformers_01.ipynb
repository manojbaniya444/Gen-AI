# Uploading file to hub

````python
from huggingface_hub import HfApi
````

````python
import os

with open("./myfile.txt") as f:
  token = f.read()
os.environ["HUGGINGFACE_TOKEN"] = token
````

````python
api = HfApi(
    token = os.getenv("HUGGINGFACE_TOKEN")
)

api.upload_file(
    path_or_fileobj="./config_folder/config.json", # our local file to upload
    path_in_repo="config.json", # path in repo
    repo_id="manojbaniya/intent-classification", # repo id
    repo_type="dataset" # repo type
)
````

## Folder

````python
from huggingface_hub import HfApi

api = HfApi(
    token = token
)

api.upload_folder(
    folder_path="./config_folder", # our local folder to upload
    repo_id="manojbaniya/intent-classification", # repo id
    repo_type="dataset", # repo type,
    path_in_repo="new" # will make a new folder
)
# .gitignore will ignore files which are need to not upload in hub
````

## A large folder

````python
api.upload_large_folder(
    folder_path="./config_folder", # our local folder to upload
    repo_id="manojbaniya/intent-classification", # repo id
    repo_type="dataset", # repo type,
)
````
