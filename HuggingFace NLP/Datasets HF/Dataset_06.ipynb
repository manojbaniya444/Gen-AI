## Pushing dataset to hub
The Hub web interface allows us to upload a dataset.

- go to `New Dataset` to create a new dataset repository.
- Pick a name and choose private or pubic dataset
- Upload a dataset and navigate to `files and versions` to add a file.
- Create Dataset Card to create a readme file for data.
- There is metadata UI with several fields to select from like license, language, and task categories and these are important tags for helping users discover out dataset on the hub.
- Now we can load dataset

```python
from datasets import load_dataset

# loading the dataset
dataset = load_dataset("manojbaniya/intent-classification")

# adding some feature
dataset = dataset.map(lambda example: {"text": example["question"], "label": example["intent"]})

# pushing to huggingface hub
dataset.push_to_hub("manojbaniya/intent-classification") # need to login with write token
```
